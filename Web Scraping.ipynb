```python
url='https://www.flipkart.com/search?q=double+door+refrigerator&sid=j9e%2Cabm%2Chzg&as=on&as-show=on&otracker=AS_QueryStore_HistoryAutoSuggest_1_1_na_na_na&otracker1=AS_QueryStore_HistoryAutoSuggest_1_1_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=double+door+refrigerator%7CRefrigerators&requestId=63396b91-ea51-4227-811d-1096194440ad&as-searchtext=d'
```

```python
permit=requests.get(url)
```

```python
permit
```

```python
t_url=BeautifulSoup(permit.text)
```

```python
t_url
```

```python
brand=t_url.select("._4rR01T")
```

```python
brand
```

```python
price=t_url.select("._1_WHN1")
```

```python
rating=t_url.select('._1lRcqv ._3LWZlK')
```

```python
review=t_url.select('._13vcmD+ span')
```

```python
import pandas as p
```

```python
table=p.DataFrame(brand,columns=['Brand'])
```

```python
table['Rating']=rating
```

```python
table['Price']=price
```

```python
table['Review']=review
```
